```python
import whisper
model = whisper.load_model("tiny", device="cuda")
```

```python
from google.colab import output
from base64 import b64decode
from io import BytesIO
from google.colab import files
from pydub import AudioSegment

def load_audio():
    uploaded = files.upload()  # Tải tệp lên
    audio_file = list(uploaded.keys())[0]  # Lấy tên tệp
    audio = AudioSegment.from_file(audio_file)  # Đọc tệp âm thanh
    return audio

def tts():
    audio = load_audio()  # Tải tệp âm thanh
    audio.export('/content/audio.wav', format='wav')  # Xuất tệp WAV

    print("Finish loading audio. Send to model")
    result = model.transcribe("/content/audio.wav", language="vietnamese")
    return result["text"]
```

```python
load_audio()
```

```python
tts()
```
